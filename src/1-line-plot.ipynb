### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the x values from 0 to 10
x = np.arange(0, 11)

# Define the y values as the square of x
y = x ** 2

# Create the plot
plt.plot(x, y)

# Add a title
plt.title("Line Plot of $y = x^2$")

# Add labels for both axes
plt.xlabel("X values")
plt.ylabel("Y values (Square of X)")

# Add a grid
plt.grid(True)

# Show the plot
plt.show()
```
