### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create the 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Data for the plots
x = np.arange(0, 11)
y = x ** 2
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]
labels_pie = ['Python', 'Java', 'C++', 'JavaScript']
sizes_pie = [40, 25, 20, 15]
colors = ['red', 'blue', 'green', 'purple']

# Line chart in the first subplot
axs[0, 0].plot(x, y)
axs[0, 0].set_title("Line Chart")
axs[0, 0].set_xlabel("X values")
axs[0, 0].set_ylabel("Y values (Square of X)")
axs[0, 0].grid(True)

# Bar chart in the second subplot
axs[0, 1].bar(categories, values, color=colors)
axs[0, 1].set_title("Bar Chart")

# Scatter plot in the third subplot
axs[1, 0].scatter(x_scatter, y_scatter)
for i, txt in enumerate(zip(x_scatter, y_scatter)):
    axs[1, 0].annotate(f'({txt[0]}, {txt[1]})', (x_scatter[i], y_scatter[i]), textcoords="offset points", xytext=(0,10), ha='center')
axs[1, 0].set_title("Scatter Plot")

# Pie chart in the fourth subplot
axs[1, 1].pie(sizes_pie, labels=labels_pie, autopct='%1.1f%%', startangle=140)
axs[1, 1].set_title("Pie Chart")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```
